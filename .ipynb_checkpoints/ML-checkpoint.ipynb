```python
df_1 = pd.DataFrame()
for i, file in enumerate(file_lst):
    user_df  = pd.read_csv(file)
    user_df  = user_df[user_df.columns[:22]][300:3100]
    lst_730  = user_df.columns[1:8].tolist()
    lst_850  = user_df.columns[8:15].tolist()
    lst_back = user_df.columns[15:22].tolist()

    for idx in range(7):
        user_df[lst_730[idx]] = user_df[lst_730[idx]] + user_df[lst_730[idx]]
        user_df[lst_850[idx]] = user_df[lst_850[idx]] + user_df[lst_850[idx]]

    uid = int(file.split('/')[-1].split('\\')[1].split(".")[0])
    user_df.insert(0,'uid',uid)
    if i<55:
        category='CN'
    elif 55<=i<=85:
        category='MCI'
    elif 85<i<=99:
        category='AD'
    user_df['Category'] = category
    df_1 = pd.concat([df_1, user_df])

uid_lst  = list(set(df_1.uid.unique().tolist()))
df_train = pd.DataFrame()
for uid in uid_lst:
    df       = df_1[df_1.uid==uid]
    df_train = pd.concat([df_train, df])
df_train = df_train.reset_index(drop=True)
del df_1
```

```python
separation_730 = [4.2,3.6,3.2,3.45,4,3.025,1]
separation_850 = [3.95,3.4,2.975,3.4,4,3,1]

od_df = pd.DataFrame()
for idx in df_train.uid.unique() : 

    uid_1   = df_train[df_train.uid==idx]
    lst_730 = df_train.columns[2:9].tolist()
    lst_850 = df_train.columns[9:16].tolist()

    for ch in range(len(lst_730)) :
        
        uid_1[lst_730[ch]] = np.log(np.mean(uid_1[lst_730[ch]]) ) - np.log(uid_1[lst_730[ch]].values ) 
        uid_1[lst_850[ch]] = np.log(np.mean(uid_1[lst_850[ch]]) ) - np.log(uid_1[lst_850[ch]].values )

        separation_by_dpf_730 = separation_730[ch] * 7
        separation_by_dpf_850 = separation_850[ch] * 7       
        
        uid_1[lst_730[ch]] = uid_1[lst_730[ch]] / separation_by_dpf_730
        uid_1[lst_850[ch]] = uid_1[lst_850[ch]] / separation_by_dpf_850
        
    od_df = pd.concat([od_df,uid_1])
```

```python
hb_df = pd.DataFrame()
for idx in od_df.uid.unique() : 

    uid_1 = od_df[od_df.uid==idx]
    col_lst = od_df.columns[2:].tolist()
    lst_730 = od_df.columns[2:9].tolist()
    lst_850 = od_df.columns[9:16].tolist()
    
    alpha_oxy_850   = 2.1838
    alpha_dox_850 = 7.0146*1.355
    alpha_oxy_730   = 10.2470
    alpha_dox_730 = 6.8501*1.355
    
    for i in range(len(lst_730)) :
        
        oxy_num = ( alpha_dox_730*uid_1[lst_850[i]] - alpha_dox_850*uid_1[lst_730[i]] )
        dox_num = ( alpha_oxy_730*uid_1[lst_850[i]] - alpha_oxy_850*uid_1[lst_730[i]] )

        oxy_denom = (alpha_dox_730*alpha_oxy_850) - (alpha_dox_850*alpha_oxy_730)
        dox_denom = (alpha_oxy_730*alpha_dox_850) - (alpha_oxy_850*alpha_dox_730)
        
        oxy = oxy_num / oxy_denom
        dox = dox_num / dox_denom
        
        uid_1[lst_730[i]] = oxy
        uid_1[lst_850[i]] = dox
        
        uid_1.rename(columns={lst_730[i] : 'hbo_{}'.format(i+1),
                              lst_850[i] : 'hbr_{}'.format(i+1)} ,inplace=True)
        
    hb_df = pd.concat([hb_df,uid_1])
```

```python
col_lst = hb_df.columns[2:-1].tolist()
```

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y


filtered_hb_df = pd.DataFrame()

for idx in hb_df.uid.unique() : 

    uid_1 = hb_df[hb_df.uid==idx]
    col_lst = uid_1.columns[2:-1].tolist()

    for j in col_lst :
        uid_1[j] = butter_bandpass_filter(uid_1[j], 0.01, 1.5, 10, order = 7)    
        
    filtered_hb_df = pd.concat([filtered_hb_df,uid_1])
```

```python
filter_df = filtered_hb_df.copy()

def lowpassfilter(signal, thresh=0.4, wavelet='dmey') :
    thresh    = thresh*np.nanmax(signal)
    coeff     = pywt.wavedec(signal, wavelet, mode = 'per')
    coeff[1:] = (pywt.threshold(i, value=thresh, mode='soft') for i in coeff[1:])
    reconstructed_signal = pywt.waverec(coeff, wavelet, mode = 'per')
    return reconstructed_signal

wav_hb_df = pd.DataFrame()

for idx in filter_df.uid.unique() : 

    uid_1 = filter_df[filter_df.uid==idx]
    col_lst = uid_1.columns[2:-1].tolist()

    for j in col_lst :
        uid_1[j] = lowpassfilter(uid_1[j], 0.3)
        
    wav_hb_df = pd.concat([wav_hb_df,uid_1])
```

```python
train   = wav_hb_df[wav_hb_df.columns[:9]]
feature = train[train.columns[3:7]].T.mean()

train_user_num = len(train.uid.unique())
train_time_seq = len(train[train.uid==1])

train_X = np.array(feature.values).reshape(train_user_num, train_time_seq)
train_Y = np.array(wav_hb_df.drop_duplicates(['uid']).Category.values) 
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder

x_train, x_test, y_train, y_test = train_test_split(train_X, train_Y, 
                                                    test_size=0.35, 
                                                    random_state=42, 
                                                    stratify=train_Y)

# scaler = RobustScaler()      # 중앙값과 사분위 값을 이용하여 데이터 변환 (이상치에 효과적) 

scaler = QuantileTransformer(output_distribution='normal')
# 1000개 분위를 사용하여 데이터를 균등분포 하도록 변환  Robust와 유사하면서 0~1사이로 데이터 압축

x_train = scaler.fit_transform(x_train)
x_test  = scaler.fit_transform(x_test)
```
